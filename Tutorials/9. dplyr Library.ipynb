## using pipe (%>%)
Explanation that %>% performs operations in a chain.

```r
# Combine multiple operations with pipes
iris %>%
  filter(Species == "setosa") %>%
  select(Sepal.Length, Sepal.Width) %>%
  mutate(Sepal.Ratio = Sepal.Length / Sepal.Width) %>%
  arrange(desc(Sepal.Ratio))
```

### grammar of data manipulation

1. filter(): Filter data based on condition.

```r
iris %>% filter(Species == "setosa", Sepal.Length > 5)
```

2. select(): Select specific columns

```r
iris %>% select(Sepal.Length, Sepal.Width)
```

3. mutate(): Create a new column or change column values

```r
iris %>% mutate(Sepal.Ratio = Sepal.Length / Sepal.Width)
```

4. summarize() and group_by(): Data summarization along with grouping

```r
iris %>%
  group_by(Species) %>%
  summarize(mean_sepal = mean(Sepal.Length))
```

5. arrange(): Data sorting

```r
iris %>% arrange(desc(Sepal.Length))
```

## Advanced Functions

```r
iris %>% distinct(Species)
```

rename(): Rename columns

```r
iris %>% rename(Sepal_Length = Sepal.Length)
```
