```r
library(ggplot2)
```

## Normal Distribution

```r
mean <- 0  
sd <- 1  
n <- 1000  

# Producing Data  
data_norm <- rnorm(n, mean = mean, sd = sd)  

# Plotting the Result
hist(data_norm, breaks = 30, main = "Normal Distribution", xlab = "Value", col = "lightblue")
```

## Bernoulli Distribution

```r
p <- 0.5
n <- 1000  

# Producing Data
data_bernoulli <- rbinom(n, 1, p)  

# Plotting the Result 
hist(data_bernoulli, breaks = 2, main = "Bernoulli Distribution", xlab = "Value", col = "lightgreen")
```

## Binomial Distribution

```r
n_trials <- 10   # Number of Trials 
p_success <- 0.5 # Probability of Success  
n <- 1000  

# Producing Data
data_binom <- rbinom(n, n_trials, p_success)  

# Plotting the Result  
hist(data_binom, breaks = 10, main = "Binomial Distribution", xlab = "Number of Successes", col = "lightcoral")
```

## Poisson Distribution

```r
lambda <- 5  # average 
n <- 1000  

# Producing Data 
data_poisson <- rpois(n, lambda)  

# Plotting the Result
hist(data_poisson, breaks = 30, main = "Poisson Distribution", xlab = "Number of Events", col = "lightyellow")
```

## Exponential Distribution

```r
rate <- 1
n <- 1000  

# Producing Data 
data_exponential <- rexp(n, rate)  

# Plotting the Result  
hist(data_exponential, breaks = 30, main = "Exponential Distribution", xlab = "Time Between Events", col = "lightpink")
```

## Gamma Distribution

```r
shape <- 2 
scale <- 2
n <- 1000  

# Producing Data 
data_gamma <- rgamma(n, shape, scale)  

# Plotting the Result 
hist(data_gamma, breaks = 30, main = "Gamma Distribution", xlab = "Value", col = "lightblue")
```

## T Distribution

```r
df <- 5    # Degree of Freedom
n <- 1000

# Producing Data
data_t <- rt(n, df)

# Plotting the Result
hist(data_t, breaks = 30, main = "t-Distribution", xlab = "Value", col = "lightblue")
```

## Chi-Square Distribution

```r
df <- 5  # Degree of Freedom 
n <- 1000  

# Producing Data  
data_chisq <- rchisq(n, df)  

# Plotting the Result 
hist(data_chisq, breaks = 30, main = "Chi-Square Distribution", xlab = "Value", col = "lightgreen")
```
