```r
ggplot(data = mpg) + geom_point(mapping = aes(y = hwy, x = displ))
# geom = geometric objects and aes = aesthetics
```

```r
ggplot(data = mpg) + geom_point(mapping = aes(y = hwy, x = displ, color = class))
# in the code above the value of color which is (class) is a feature that helps the scatter plot 
# to divide each model of car. but also you can give it a specific color
```

```r
ggplot(data = mpg, aes(x = displ, y = hwy, color = class, shape = drv)) +
  geom_point(size = 3) +
  labs(title = "Scatter Plot of Engine Displacement vs Highway MPG",
       x = "Engine Displacement (L)",
       y = "Highway MPG")
```

```r
ggplot(data = mpg) + geom_smooth(mapping = aes(y = hwy, x = displ))
```

```r
ggplot(data = mpg) + geom_smooth(mapping = aes(y = hwy, x = displ, color = class))
```

```r
ggplot(data = mpg) + geom_point(mapping = aes(y = hwy, x = displ, color = drv)) + 
  geom_smooth(mapping = aes(y = hwy, x = displ, color = drv, linetype = drv))
# the statement below is completely similar to the above one, because of 
#the inheritance feature made in ggplot function
```

```r
ggplot(data = mpg, mapping = aes(y = hwy, x = displ, color = drv)) +
  geom_point() + 
  geom_smooth(mapping = aes(linetype = drv))
# you don't have to use data argument in ggplot and if you work with different and
#variant datasets, you can add those seperately in each geometric object
```

```r
ggplot(data = mpg, aes(x = displ, y = hwy)) +
  geom_point() +
  labs(title = "Scatter Plot of Engine Displacement vs Highway MPG",
       x = "Engine Displacement (L)",
       y = "Highway MPG")
```

```r
mpg_avg <- aggregate(hwy ~ year, data = mpg, FUN = mean)

ggplot(data = mpg_avg, aes(x = year, y = hwy)) +
  geom_line() +
  labs(title = "Average Highway MPG per Year",
       x = "Year",
       y = "Highway MPG")
```

```r
ggplot(data = mpg, aes(x = displ, y = hwy)) +
  geom_point() +
  geom_smooth(method = "lm", se = FALSE, color = "blue") +
  labs(title = "Engine Displacement vs Highway MPG with Trend Line",
       x = "Engine Displacement (L)",
       y = "Highway MPG")
```

```r
ggplot(data = mpg, aes(x = class)) +
  geom_bar() +
  labs(title = "Number of Vehicles in Each Class",
       x = "Vehicle Class",
       y = "Count")
```

```r
ggplot(data = mpg, aes(x = hwy)) +
  geom_histogram() +
  labs(title = "Distribution of Highway MPG",
       x = "Highway MPG",
       y = "Frequency")
```

```r
ggplot(data = mpg, aes(x = hwy)) +
  geom_histogram(binwidth = 2, fill = "skyblue", color = "black") +
  labs(title = "Distribution of Highway MPG",
       x = "Highway MPG",
       y = "Frequency")
```

```r
ggplot(data = mpg, aes(x = class, y = hwy)) +
  geom_boxplot(fill = "lightgreen") +
  labs(title = "Box Plot of Highway MPG by Vehicle Class",
       x = "Vehicle Class",
       y = "Highway MPG")
```

```r
ggplot(data = mpg, aes(x = displ, y = hwy, color = class)) +
  geom_point(size = 2) +
  geom_smooth(method = "lm", se = FALSE) +
  theme_minimal() +
  labs(title = "Enhanced Scatter Plot with Minimal Theme",
       x = "Engine Displacement (L)",
       y = "Highway MPG") +
  theme(plot.title = element_text(hjust = 0.5))
```

```r
my_df <- diamonds
ggplot(data = my_df) +
  geom_bar(aes(x = cut)) # here a statistical transformation happens and make a new dataset
#will make, because the y axis is empty and R itself work with it
# statistical transformation use a function to do this, for example geombar uses stat_count as default
```

```r
ggplot(my_df) +
  geom_bar(aes(x = cut, y = shape))
# it returns an error because geombar need to use stat count
```

```r
ggplot(my_df) +
  stat_count(aes(x = cut)) # the result is similar as the statement above
####### an example when we have the counts
```

```r
a_df <- data.frame(cut = c("Fair", "Good", "Very Good", "Premium", "Ideqal"),
                   count = c(100, 150, 200, 130, 400))
```

```r
ggplot(a_df) +
  geom_bar(aes(x = cut, y = count)) # it returns an error because y axis should not mention
#and stat count do it itself
```

```r
#what is the solution?
ggplot(a_df) +
  geom_bar(aes(x = cut, y = count), stat = "Identity") # identity is a statistical transformation which do nothing
```

```r
ggplot(a_df) +
  geom_bar(aes(x = cut)) # please check the result
```

```r
?geom_col #is similar to geom bar but the statistical function is identity
ggplot(a_df) +
  geom_col(aes(x = cut, y = count))
```
