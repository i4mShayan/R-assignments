## الف

```r
dice <- function(n) {
  results <- sample(1:6, size = n, replace = TRUE)
  freq <- table(results) / n * 100
  barplot(freq, main = paste("Throws:", n), xlab = "Dice Face", ylab = "Frequency (%)", col = "black", ylim = c(0, 30))
}

par(mfrow = c(1, 3))
dice(100)
dice(1000)
dice(10000)
```

## ب

```r
set.seed(42)

par(mfrow = c(1, 3))
dice(100)
dice(1000)
dice(10000)
```
