# Part A

```r
set.seed(42)
# random_numbers <- rexp(n = 100, rate = 0.2)
# data_frame <- data.frame(Values = random_numbers)

random_numbers <- sample(1:100, size = 250, replace = TRUE)
data_frame <- data.frame(Values = random_numbers)

hist(
  data_frame$Values,
  breaks = 10,
  col = "blue",
  main = "Distribution with Density Curve",
  xlab = "Values",
  ylab = "Frequency",
  probability = TRUE
)
lines(
  density(data_frame$Values),
  col = "red",
  lwd = 2
)

mean_value <- mean(data_frame$Values)
std_dev <- sd(data_frame$Values)
value_range <- range(data_frame$Values)

cat("Mean:", mean_value, "| Standard Deviation:", std_dev, "| Range: from", value_range[1], "to", value_range[2])
```

# Part B

```r
normalized_values <- (data_frame$Values - min(data_frame$Values)) / (max(data_frame$Values) - min(data_frame$Values))

data_frame$Normalized_Values <- normalized_values

hist(
  data_frame$Normalized_Values,
  breaks = 10,
  col = "blue",
  main = "Normalized Distribution",
  xlab = "Normalized Values",
  ylab = "Frequency",
  probability = TRUE
)
lines(
  density(data_frame$Normalized_Values),
  col = "red",
  lwd = 2
)

mean_normalized <- mean(data_frame$Normalized_Values)
std_dev_normalized <- sd(data_frame$Normalized_Values)
range_normalized <- range(data_frame$Normalized_Values)

cat("Mean:", mean_normalized, "| Standard Deviation:", std_dev_normalized, "| Range: from", range_normalized[1], "to", range_normalized[2])
```

```r
standardized_values <- (data_frame$Values - mean(data_frame$Values)) / sd(data_frame$Values)

data_frame$Standardized_Values <- standardized_values

hist(
  data_frame$Standardized_Values,
  breaks = 10,
  col = "blue",
  main = "Standardized Distribution",
  xlab = "Standardized Values (z-scores)",
  ylab = "Frequency",
  probability = TRUE
)
lines(
  density(data_frame$Standardized_Values),
  col = "red",
  lwd = 2
)

mean_standardized <- mean(data_frame$Standardized_Values)
std_dev_standardized <- sd(data_frame$Standardized_Values)
range_standardized <- range(data_frame$Standardized_Values)

cat("Mean:", mean_standardized, "| Standard Deviation:", std_dev_standardized, "| Range: from", range_standardized[1], "to", range_standardized[2])
```

# توضیح  
### اهمیت نرمال‌سازی داده‌ها
نرمال‌سازی داده‌ها به ما این امکان را می‌دهد که مقادیر با مقیاس‌ها و اندازه‌های متفاوت را با یکدیگر قابل مقایسه کنیم. به‌عنوان مثال، اگر یکی از متغیرها در بازه 0 تا 1000 باشد و دیگری در بازه 0 تا 10، متغیری که مقادیر بزرگ‌تری دارد تاثیر بیشتری در تحلیل خواهد داشت. نرمال‌سازی با یکسان کردن مقیاس داده‌ها، این مشکل را برطرف کرده و تحلیل را دقیق‌تر و منصفانه‌تر می‌کند.

### مزایا و معایب Range Normalization:
- مزایا:
  - داده‌ها را به یک بازه مشخص (معمولا 0 تا 1) می‌برد که مقایسه داده‌ها را ساده‌تر می‌کند.
  - ترتیب نسبی مقادیر داده‌ها را حفظ می‌کند.
  - محاسبه آن ساده و سریع است.

- معایب:
  - به مقادیر پرت حساس است؛ یک مقدار بسیار بزرگ یا کوچک می‌تواند کل مقیاس را تحت تاثیر قرار دهد.
  - با اضافه شدن داده‌های جدید، نیاز به نرمال‌سازی مجدد وجود دارد.

### مزایا و معایب Distribution Normalization یا همان استاندارد سازی:
- مزایا:
  - داده‌ها را استاندارد می‌کند (میانگین = 0 و انحراف معیار = 1)، که این کار برای روش‌های آماری مانند PCA یا الگوریتم‌های مبتنی بر فاصله بسیار مفید است.
  - تأثیر متغیرهای مختلف را یکسان می‌کند، حتی اگر مقیاس‌های متفاوتی داشته باشند.

- معایب:
  - نسبت به روش Range کمی پیچیده‌تر است، چرا که نیاز به محاسبه میانگین و انحراف معیار دارد.
  - همچنین اعداد اسکیل حقیقی خود را در این روش از دست می‌دهند.
